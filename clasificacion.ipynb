```python
# Dividir las variables en categóricas y numéricas
categorical_columns = ["Regiones", "NivelDeRiqueza", "ViolenciaEmocional",
                       "NivelDeEducacion"]  # No debes colocar el Target aquí

numerical_columns = ['Edad', 'EdadPrimNac']

prep = ColumnTransformer([('encoder', OneHotEncoder(), categorical_columns),
                           ('scaler', StandardScaler(),numerical_columns),],
                          remainder= 'drop')
```

```python
#rf_model = RandomForestClassifier(n_estimators=200,
#                                  random_state=42,
#                                  max_depth=10,
#                                  min_samples_split=10,
#                                  min_samples_leaf=5,
#                                  max_features="sqrt",
#                                  n_jobs=-1,
#                                  class_weight="balanced")
```

```python
from sklearn.pipeline import Pipeline
rf_pip = Pipeline([
    ('preproc', prep),
    ('rf', RandomForestClassifier(n_estimators=200,
                                  random_state=42,
                                  max_depth=10,
                                  min_samples_split=10,
                                  min_samples_leaf=5,
                                  max_features="sqrt",
                                  n_jobs=-1,
                                  class_weight="balanced"))
])
```

```python
rf_pip.fit(X_train, y_train)
```

```python
from sklearn.inspection import permutation_importance
import multiprocessing
```

```python
importancia = permutation_importance(
                estimator    = rf_pip,##modelo entrenado
                X            = X_train,
                y            = y_train,
                n_repeats    = 5,
                scoring      = 'roc_auc',##Por tener clases desbalanceadas, evalua como el modelo distingue entre 2 clases
                n_jobs       = multiprocessing.cpu_count() - 1,
                random_state = 123
             )

# Se almacenan los resultados (media y desviación) en un dataframe
df_importancia = pd.DataFrame(
                    {k: importancia[k] for k in ['importances_mean', 'importances_std']}
                 )
df_importancia['feature'] = X_train.columns
df_importancia.sort_values('importances_mean', ascending=False)
```

```python
from sklearn import metrics
accuracy_train_RF = metrics.accuracy_score(y_train, rf_pip.predict(X_train))
accuracy_test_RF = metrics.accuracy_score(y_test, rf_pip.predict(X_test))
print("El accuracy train es rf:", accuracy_train_RF)
print("El accuracy test es rf:", accuracy_test_RF)
```
